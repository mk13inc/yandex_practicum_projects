**Название проекта**: Исследование рынка компьютерных игр

**Описание проекта**: Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, *Xbox* или *PlayStation*). Вам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.
Перед вами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и вы планируете кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируете ли вы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.
В наборе данных попадается аббревиатура *ESRB* (*Entertainment Software Rating Board*) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. **Цель исследования**: Найти закономерности, определяющие успешность компьютерной игры для разработки потенциально популярного продукта и планирования рекламной кампании

### Загрузка данных из файла в датафрейм

```python
import pandas as pd  # Импорт библиотеки pandas
import matplotlib.pyplot as plt  # Импорт библиотеки для работы с графиками

# открытие датасета:
data = pd.read_csv('/datasets/games.csv')

# переменные для задания цвета шрифта при выводе информации:
green, red, blue, pink, black, bold, end = '\33[92m', '\33[91m', '\33[36m', '\33[95m','\33[7m', '\33[1m','\33[0m'

# обращение к встроенной цветовой палитре plt.colormaps['tab10'].colors 
# в тренажере выдает ошибку, поэтому задаю палитру вручную:
color_list = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', 
               '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']

# Для упрощения оформления и вывода информации использую вспомогательные функции:
def draw_bar(data_set, y, title, xlabel, ylabel='Региональные продажи, млн. копий'):
    """
    Функция строит столбчатый график для указанного датасета
    """
    data_set.plot(kind='bar', y=y, color=color_list, figsize=(20, 10), rot=0)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    # Добавление горизонтальной линии для среднего значения с красным цветом и пунктирной линией
    plt.axhline(y=data_set[y].describe()['mean'], label='mean', color='r', linestyle='--')
    # Добавление горизонтальной линии для медианного значения с синим цветом и точечной линией
    plt.axhline(y=data_set[y].describe()['50%'], label='median', color='b', linestyle=':')
    plt.grid(alpha=0.5, linestyle='--', visible=True)
    plt.legend()
    plt.show()

def draw_barh(data_set, y, title, ylabel, xlabel='Региональные продажи, млн. копий', color=color_list):
    """
    Функция строит горизонтальный столбчатый график для указанного датасета
    """
    data_set.plot(y=y, kind='barh', figsize=(20, 10), color=color_list);
    # Установка заголовка графика
    plt.title(title)
    # Добавление подписи для оси Y
    plt.ylabel(ylabel)
    # Добавление подписи для оси X
    plt.xlabel(xlabel)
    # Добавление вертикальной линии для среднего значения с красным цветом и пунктирной линией
    plt.axvline(x=data_set[y].describe()['mean'], label='mean', color='r', linestyle='--')
    # Добавление вертикальной линии для медианного значения с синим цветом и точечной линией
    plt.axvline(x=data_set[y].describe()['50%'], label='median', color='b', linestyle=':')
    plt.grid(alpha=0.5, linestyle='--', visible=True)
    plt.legend()
    plt.show()

def draw_pie(data_set, y, title):
    """
    Функция для построения диаграммы рассеяния
    """
    data_set[y].plot(kind='pie', title=title, figsize=(10, 10), autopct='%1.1f%%')
    plt.gca().axes.yaxis.set_visible(False) # скрытие заголовка оси X
    plt.show()

def draw_scatter(data_set, x, y, title, xlabel, xlim, figsize=(20, 10), ylabel='Объем продаж, млн. копий'):
    """
    Функция для построения круговой диаграммы для указанного датасета
    """
    data_set.plot(x=x, y=y, figsize=figsize, alpha=0.3, 
    title=title, xlabel=xlabel, ylabel=ylabel, style='o', grid=True, xlim=xlim);
    plt.grid(alpha=0.5, linestyle='--', visible=True)
    plt.show()

def describe(values, col):
    """
    Функция для вывода описательных характеристик набора
    числовых данных на основе информации метода describe:
    медиана, межквартильный размах, границы нормальных значений
    """
    d = values[col].describe()
    Q1, Q2, Q3 = d['25%'], d['50%'], d['75%']
    IQR = Q3 - Q1
    print(f"Медианное значение: {Q2:.3f}")
    print(f"Межквартильный размах: от {Q1:.3f} до {Q3:.3f}, равен {IQR:.3f} ")
    print(f"Границы нормальных значений: от {max(Q1 - 1.5*IQR, 0)} до {Q3 + 1.5*IQR:.3f}")
    
def kendall(dataset, x, y):
    """
    kendall(dataset, x, y)
    Функция выводит величину коэффициента 
    корреляции Кендалла для наборов x и y.
    Обычно, корреляция Кендалла предпочтительнее корреляции Спирмена 
    из-за меньшей чувствительности к грубой ошибке (более надежная) 
    и меньшей асимптотической дисперсии (более эффективная)
    """
    return round(dataset[[x, y]].corr(method='kendall')[y][0], 3)
    
def is_that_snake(list_of_strings):
    """
    Данная функция получает на входе список со строковыми значениями (например, список заголовков датафрейма)
    и поочередно проверяет наличие НЕжелательных для оформления символов.
    Желательные символы: английский алфавит в нижнем регистре, нижнее подчеркивание и цифры от 0 до 9
    """
    snake_symbols, bad_list = list('abcdefghijklmnopqrstuvwxyz_0123456789'), []
    for i in list_of_strings:
        if set(i).issubset(snake_symbols): # проверка, что все элементы подмножества i
            pass                           # принадлежат множеству snake_symbols
        else:
            bad_list.append(i)
    return f'(!) Следует обратить внимание: {bad_list}' if bad_list else '✅ Проблем с регистром не обнаружено'    

pd.options.display.float_format = ("{:.2f}".format)  # формат вывода float чисел датафрейма
```

* заполнение медианными значениями не подходит, потому что для разных платформ игра может иметь довольно разные оценки;

Поэтому, записи со значением `tbd` будем также трактовать, как записи без оценки.

Попробуем немного исследовать динамику продаж. 

В таблице представлены данные о продажах игровых платформ по годам:

```python
sale_dynamic = data_actual.pivot_table(index='year_of_release', columns='platform', 
                                       values='total_sales', aggfunc='sum')
sale_dynamic
```

Я задал вспомогательную функцию, которая поможет оценить характер изменения значений объема прибыли:

```python
def dynamic_check(cols):
    """
    Функция оценки характера изменения значений
    для числовых списков размером 3
    """
    up_3, down_3, up_2, down_2, unknown = [],[],[],[],[]
    for platform in cols:
        sales_list = sale_dynamic[platform].to_list()      # формируем список из значений продаж по годам, 
                                                           # например [43.76, 27.78, 15.14]
        if sales_list[0] >= sales_list[1] >= sales_list[2]:   # проверка условия соотношения
            down_3.append([platform, sales_list])             # величин значений из списка
            
        elif sales_list[0] <= sales_list[1] <= sales_list[2]: #
            up_3.append([platform, sales_list])               #
            
        elif sales_list[1] >= sales_list[2]:                  # 
            down_2.append([platform, sales_list])             # 
            
        elif sales_list[1] <= sales_list[2]:                  # 
            up_2.append([platform, sales_list])               #
            
        else:                                                 #
            unknown.append([platform, sales_list])            #
            
    if down_3:
        print(red + '😩 Платформы, которые показали снижение продаж в течение всего периода:' + end)
        for el in down_3:
            print('-', el[0], f'{el[1]}')
    
    if down_2:
        print(pink + '😕 Платформы, которые показали снижение продаж за последние 2 года:' + end)
        for el in down_2:
            print('-', el[0], el[1])

    if up_3:
        print(green + 'Платформы, которые показали рост продаж в течение всего периода:' + end)
        for el in up_3:
            print('-', el[0], el[1])

    if up_2:
        print(green + 'Платформы, которые показали рост продаж за последние 2 года:' + end)
        for el in up_2:
            print('-', el[0], el[1])
    
    if unknown:
        print('🤨 Платформы, у которых неразборчивое поведение:' + end)
        for el in unknown:
            print('-', el[0], el[1])
```

Применим функцию к имеющимся данным:

```python
dynamic_check(sale_dynamic.columns)
```

**Вывод:** Судя по полученным данным, все платформы показали 'отрицательный рост'. Тем не менее, две игровые платформы - **PS4** и **XOne** хоть и показали снижение продаж за последние два года, но даже в этом контексте, они лидеры по продажам за весь исследуемый период и за 2016 год отдельно.

```python
# формирование сводной таблицы с распределением игр по жанрам:
genre_spread = data_actual.pivot_table(index='genre', values='name', aggfunc=lambda x: len(x.unique()))

# добавление столбика с долей игр от всего объема:
genre_spread['rate'] = round(100 * genre_spread['name'] / genre_spread['name'].sum(), 1)

# изменение заголовков:
genre_spread.columns = ['games_per_genre', 'rate, %']

# сортировка значений:
genre_spread.sort_values(by='games_per_genre', ascending=False)
```

Можно отобразить распределение в виде круговой диаграммы:

```python
draw_pie(data_set=genre_spread, y='games_per_genre', title='Распределение игр по жанрам')
```

Поскольку в исследуемом датасете напрямую нет информации о *суммах продаж*, а есть только данные о количестве проданных копий, то оценить **прибыльность** можно рассчитав отношение <ins>суммарного объема продаж</ins> к <ins>количеству игр</ins> в пределах жанра. Иными словами, найти среднее значение проданных копий на одну игру. При этом следует сделать допущение, что цена продажи всех исследуемых игр одинакова.

Сформирую сводную таблицу, содержащую данные о количестве игр `games_per_genre`, объемах продаж `total_sales` и рассчитаю среднее значение проданных копий на одну игру `copies`:

```python
genre_copies = data_actual.pivot_table(index='genre', 
                                       values=['name', 'total_sales'], 
                                       aggfunc={'name': lambda x: len(x.unique()), 'total_sales': "sum"})
genre_copies['copies'] = round(genre_copies['total_sales'] / genre_copies['name'], 3)
genre_copies.columns = ['games_per_genre', 'total_sales', 'copies']
genre_copies.sort_values(by='copies', ascending=False)
```

Для более наглядного изучения влияния рейтинга на продажи игр, я введу дополнительное числовое обозначение категорий в зависимости от возрастных ограничений на основе [информации имеющейся в открытом доступе](https://ru.wikipedia.org/wiki/Entertainment_Software_Rating_Board). 

Продублирую градацию:

Код рейтинга | Описание | Числовое обозначение
----|----|---
«E» («Everyone») | «Для всех»: Содержание вполне подходит для всех возрастов | 0
«E10+» («Everyone 10 and older») | «Для всех от 10 лет и старше» | 10
«T» («Teen») | «Подросткам»: Игра подходит для лиц от 13 лет | 13
«M» («Mature») | «Для взрослых»: Материалы игры не подходят для лиц младше 17 лет | 17

Для добавления обозначений использую вспомогательную функцию `rating_code`:

```python
def rating_code(row):
    """
    Функция для присвоения числового обозначения
    в зависимости от категории рейтинга
    """
    return {'E':0, 'E10+':10, 'T':13, 'M':17, 'no rating':100}[row]
```

Формирую сводную таблицу по региональным объемам продаж в разбивке по категориям рейтинга.  
Добавляю в таблицу столбец `age` с числовым обозначением рейтинга.

```python
na_sales_rating = data_actual.pivot_table(index='rating', values='na_sales', aggfunc='sum')
na_sales_rating['age'] = na_sales_rating.index.to_series().apply(rating_code)
na_sales_rating
```

Во-первых, улучшение технологий позволило создавать более сложные и интересные игры. Во-вторых, увеличение числа игроков привело к росту спроса на новые игры. В-третьих, развитие интернета и онлайн-игр также способствовало увеличению популярности видеоигр. Однако стоит отметить, что с наступлением 2010-х гг. наблюдается некоторое снижение количества выпускаемых игр. Это может быть связано с насыщением рынка или с изменением предпочтений потребителей.
* Произведена оценка объема продаж игр по платформам: самый высокий показатель у **PS2**: 1255.8 млн. проданных копий. Определено время жизни игровых платформ: как правило, платформа существует немного больше 10 лет.
* Для формирования прогноза закономерностей, определяющих успешность игры, на 2017 год был выбран актуальный период исследования данных: 2014-2016.

На основании данных имеющихся в распоряжении за актуальный выбранный период времени был проведен анализ и получены следующие результаты:
1. При рассмотрении игровой отрасли в глобальном плане:
* **Потенциально прибыльные платформы**: в общей картине самыми высокими показателями по продажам обладают игровые платформы **PS4**: 288.1 млн проданных копий за период исследования и **XOne**: 140.4 млн. копий. Данные платформы также обладают большим списком игр с высокоми показателями продаж.
* **Влияние отзывов на объем продаж**: как правило, оценка критиков демонстрирует слабую положительную корреляцию (коэффициент корреляции ~0.35) с объемом продаж, в то время как оценка пользователей практически не оказывает влияния (коэффициент корреляции ~0.0). Такое распределение наблюдается как внутри одной популярной игровой платформы (PS4), так и среди других платформ.
* **Распределение игр по жанрам**: жанр с самым высоким показателем выпущенных игр - **Action**: 338 игр за исследуемый период. В то же время, по показателю среднего значения проданных копий на одну игру лидирует жанр **Shooter**: среднее значение составляет 3.35 млн. А **Action** на 8-м месте с 0.59 млн. копий. Хуже всего с продажами у жанра **Strategy** - 0.12 млн. копий.
2. При рассмотрении локальных секторов игровой отрасли:  
 Составлены портреты типичного пользователя для исследумых регионов (Северная Америка, Европа и Япония):
* типичный пользователь из Северной Америки:
  * предпочитает игровые платформы **PS4** и **XOne**
  * больше всего любит жанры **Shooter** и **Action**
  * как правило играет в игры с рейтингом **M** (+17)
* типичный пользователь из Европы:
  * предпочитает играть на **PS4**
  * больше всего любит жанры **Action** и **Shooter**
  * как правило играет в игры с рейтингом **M** (+17)
* типичный пользователь из Японии:
  * предпочитает играть на **Nintendo 3DS**
  * больше всего любит жанры **Role-Playing** и **Action**
  * как правило играет в игры без рейтинга ESRB
* Вывод о портрете пользователей можно дополнить информацией о том, что для сектора Северной Америки и Европы наиболее популярны как игры имеющие глобальную популярность, так и локальные. В то время как в Японии явное преобладание популярности локальных игр.
* Резюмируя информацию о портретах типичного пользователя, можно утверждать, что для выхода на пользовательский рынок Северной Америки и Европы можно принять единую стратегию и руководствоваться едиными факторами: портрет пользователей этих регионов, за исключением небольшой разницы в показателях, отличаются незначительно. Выход на пользователей из Японии требует выбора отдельной стратегии и факторов - их предпочтения существенно отличаются от других регионов.
